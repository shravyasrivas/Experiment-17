```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
data = {
    'Days': ['Mon', 'Tue', 'Wed', 'Thu', 'Fri'],
    'Study_Hours': [2, 3, 2.5, 4, 3.5],
    'Marks': [50, 60, 55, 70, 65],
    'Attendance': [60, 70, 65, 80, 75],
    'Sleep_Hours': [6, 7, 6, 8, 7],
    'Assignments_Completed': [1, 2, 1, 3, 2]
}
df = pd.DataFrame(data)
display(df)
```

```python
#Line Chart(Trend)
plt.plot(df['Days'],df['Study_Hours'],marker='p', color='gold') #used p instead of o
plt.title(' Shravya Study Hour Trend')
plt.xlabel('Days')
plt.ylabel('Study Hours')
plt.show()
```

```python
#advanced line chart
plt.figure(figsize=(7,4))
plt.plot(df['Days'],df['Study_Hours'],
         marker='o', color='steelblue',label='Study Hours')
plt.plot(df['Days'],df['Marks'],
         marker='s',color='darkorchid',label='Marks')
plt.xlabel('Days')
plt.ylabel('Values')
plt.title('Shravya Study Hours and Marks')
plt.legend()
plt.show()
#* star s square #k black, r red, b blue g green
```

```python
#bar chart
plt.bar(df['Days'],df['Marks'], color='grey',edgecolor='black')
plt.xlabel('Days')
plt.ylabel('  Marks')
plt.title('Shravya Study Days and Marks')
plt.grid(axis='y' , alpha=0.67, color='firebrick', linestyle='--') #alpha is used for intensity lies btw 0 to 1 1 is amx 0 is min
plt.show()
```

```python
#advanced bar graph
bars=plt.bar(df['Days'],df['Marks'],color='sienna',edgecolor='khaki')
for bar in bars:
  y= bar.get_height()
  plt.text(bar.get_x()+ bar.get_width()/2,y, str(y),
           ha='center', va='bottom') #coverts y into string
plt.xlabel('Days')
plt.ylabel('  Marks')
plt.title('Shravya Study Days and Marks')
plt.grid(axis='y', color='black', linestyle='--',alpha=0.4)#adds horizontal grid for readability
plt.show()

```

```python
import numpy as np
import matplotlib.pyplot as plt
x= np.arange(len(df['Days']))#possitions of bars
width= 0.36 #width od each bar
plt.bar(x-width/2,df['Study_Hours'],width,label='Study Hours',color='palegoldenrod')
plt.bar(x+width/2,df['Marks'],width,label='Marks', color='palevioletred')
plt.xticks(x,df['Days']) #to label days
plt.xlabel('Days')
plt.ylabel('Study Hours')
plt.title('Shravya Study Days and Marks')
plt.show()
```

```python
#histogram- frequency
import matplotlib.pyplot as plt
import numpy as np
#example data(more value btr the histogram)
marks=[50,60,55,70,65,75,80,85,90,45,68,72,77,82,88]
#calculating mean
mean_value=np.mean(marks)
plt.figure(figsize=(7,4))
plt.hist(marks,bins=3,color='salmon',edgecolor='olive') #higher limt wont be counted
#mean line
plt.axvline(mean_value,color='mediumorchid',linestyle='--',linewidth=2,label='Mean')
plt.xlabel('Marks')
plt.ylabel('Frequency')
plt.title('Shravya Marks Histogram')
plt.legend()
plt.show()
```

```python
#scatter plot(relationship)
import matplotlib.pyplot as plt
plt.scatter(df['Study_Hours'],df['Marks'], marker='*')
plt.title("Shravya Study Hours vs Marks")
plt.xlabel("Study Hours")
plt.ylabel("Marks")
plt.show()
#seaborn directly converts
```

```python
#conditional scatter plot
#example category
df['Result']=['Fail','Pass','Fail','Pass','Pass']
colors=['tomato' if r=='Fail' else 'palegreen' for r in df['Result']]
plt.scatter(df['Study_Hours'],df['Marks'],c=colors)
plt.title("Shravya Study Hours vs Marks")
plt.xlabel("Study Hours")
plt.ylabel("Marks")
plt.show()
#red fail green pass
```

```python
#Dataset 2 to undersatnd diff btw seaborn and matplotlib
data = {
    'Day': ['Mon','Tue','Wed','Thu','Fri','Sat','Sun','Mon','Tue','Wed'],
    'Region': ['North','South','East','West','North','South','East','West','North',
               'South'],
    'Sales': [200,220,250,270,300,350,400,280,260,290],
    'Profit': [20,25,30,35,50,60,80,40,38,45],
    'Customers': [30,35,40,45,50,65,80,55,48,52],
    'Category': ['Food','Clothing','Electronics','Food','Clothing','Electronics',
                 'Food','Clothing','Electronics','Food']
}
df = pd.DataFrame(data)
display(df)
```

Same Graphs using Seaborn (btr Visualization) (no label)

```python
#line plot
sns.lineplot(x='Day',y='Sales',data=df,errorbar=None)
plt.title('Shravya')
plt.show()
```

```python
#barchart
#Bar plot
sns.barplot(x='Day', y='Sales', data = df,color='maroon')
plt.title('Shravya')
plt.xlabel('Day')
plt.ylabel('Sales')
plt.show()
```

```python
#histogram
sns.histplot(df['Sales'], bins=5,color='coral')
plt.title("Shravya")
plt.show()
```

```python
#scatter plot
sns.scatterplot(x='Sales',y='Profit',data=df)
plt.title('Shravya')
plt.show()
```
